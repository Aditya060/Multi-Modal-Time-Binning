```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv('sorted_file(Product_2).csv')
```

```python
dates = pd.to_datetime(df['Processed_Date'])
ratings = df['Reviews']
```

```python
df['Processed_Date'] = pd.to_datetime(df['Processed_Date'])
df.head()
def plot():
#     rating_data = df
#     rating_data = df[df['Ratings']]
    rating_data = df.resample('M', on = 'Processed_Date').count()
#     print(rating_data)
    
    plt.figure(figsize = (10,6))
    plt.plot(rating_data.index, rating_data['Ratings'],marker = 'o', linestyle = ' ')
    plt.xlabel('Date')
    plt.ylabel('Number of Ratings')
    plt.title(f'Rating Over Time')
    plt.grid(True)
    plt.show()
```

```python
plot()
```

```python
df['Processed_Date'] = pd.to_datetime(df['Processed_Date'])
def plot2(rating):
    rating_data = df[df['Ratings'] == rating ]
    rating_data = rating_data.resample('M', on = 'Processed_Date').count()
    
    plt.figure(figsize = (10,6))
    plt.plot(rating_data.index, rating_data['Ratings'], marker = 'o', linestyle = ' ')
    plt.xlabel('Date')
    plt.ylabel('Number of Ratings')
    plt.title(f'Rating {rating} Over Time')
    plt.grid(True)
    plt.show()
```

```python
plot2(1)
```

```python
plot2(2)
```

```python
df['Processed_Date'] = pd.to_datetime(df['Processed_Date'])

def plot2():
    plt.figure(figsize = (10,6))
    for rating in range(1,6):
       
        rating_data = df[df['Ratings'] == rating ]
        rating_data = rating_data.resample('M', on = 'Processed_Date').count()

        
        plt.plot(rating_data.index, rating_data['Ratings'], marker = '.', linestyle = '-',  label = 'Rating {rating}'.format(rating = rating))
        plt.xlabel('Date')
        plt.ylabel('Number of Ratings')
        plt.title(f'Rating Over Time')
        plt.grid(True)
        plt.legend()
#         plt.legend([1,2,3,4,5])
    plt.show()
```

```python
plot2()
```
